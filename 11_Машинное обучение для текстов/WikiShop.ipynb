Теперь проведем лемматизацию текстов и избавимся от лишних символов с помощью регулярных выражений:

```python
def lemmatize(text):
    def get_wordnet_pos(word):
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {
            "J": wordnet.ADJ,
            "N": wordnet.NOUN,
            "V": wordnet.VERB,
            "R": wordnet.ADV,
        }
        return tag_dict.get(tag, wordnet.NOUN)

    import nltk
    from nltk.corpus import wordnet
    from nltk.stem import WordNetLemmatizer

    nltk.download("wordnet")

    lemmatizer = WordNetLemmatizer()
    word_list = nltk.word_tokenize(text)
    lemmatized_output = " ".join(
        [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in word_list]
    )
    return lemmatized_output


def clear_text(text):
    text = re.sub(r"[^a-zA-Z ]", " ", text)
    text = text.split()
    text = " ".join(text)
    return text
```

```python
try:
    df = pd.read_csv("lem_data.csv")
    print("Обработанные данные загружены из файла.")
    display(df.head())
except:
    df["text"] = df["text"].progress_apply(clear_text)
    print("Текст после регулярных выражений:")
    display(df.head())
    df["text"] = df["text"].parallel_apply(lemmatize)
    print("Текст после лемматизации:")
    display(df.head())
    df["text"] = df["text"].str.lower()
    print("Текст после перевода в нижний регистр:")
    display(df.head())
```

## Обучение моделей

Попробуем два подхода:
* эмбеддинг с помощью готовой модели ***Bert***
* векторизация с помощью метрики ***TF-IDF***

### BERT

Уменьшим для работы с BERT количество данных до 2000 строк:

```python
df_bert = df.head(2000).reset_index(drop=True)
```

#### Эмбеддинг

Перед тем как начать обучать модели необходимо создать матрицу признаков из эмбеддингов. Для этого воспользуемся моделью ***BERT***. 

Сперва токенизируем наши данные:

```python
print("Tokenizing...")
tokenizer = transformers.AutoTokenizer.from_pretrained("unitary/toxic-bert")
tokenized = df_bert["text"].progress_apply(
    lambda x: tokenizer.encode(
        x, truncation=True, add_special_tokens=True, max_length=512
    )
)
print("Done!")
```

Поиск максимального количества токенов:

```python
max_len = 0
for i in tqdm(tokenized.values, desc="Max length"):
    if len(i) > max_len:
        max_len = len(i)
print(max_len)
```

Padding, чтобы длины данных в корпусе были равными, и создание маски:

```python
padded = np.array(
    [i + [0] * (max_len - len(i)) for i in tqdm(tokenized.values, desc="Padding")]
)

attention_mask = np.where(padded != 0, 1, 0)
print(attention_mask.shape)
```

Инициализируем саму модель класса BertModel:

```python
# код на случай скачанных моделей Bert
# config = transformers.BertConfig.from_json_file("bert_config.json")
# model = transformers.BertForPreTraining.from_pretrained(
#    "bert_model.ckpt.index", from_tf=True, config=config
# )
model = transformers.AutoModel.from_pretrained("unitary/toxic-bert")
model = model.to("cuda")
print("Done")
```

Преобразуем данные в эмбеддинги:

```python
batch_size = 2
embeddings = []
for i in tqdm(range(padded.shape[0] // batch_size)):
    batch = torch.cuda.LongTensor(padded[batch_size * i : batch_size * (i + 1)])
    attention_mask_batch = torch.cuda.LongTensor(
        attention_mask[batch_size * i : batch_size * (i + 1)]
    )

    with torch.no_grad():
        batch_embeddings = model(batch, attention_mask=attention_mask_batch)

    embeddings.append((batch_embeddings[0][:, 0, :]).cpu().numpy())
```

```python
features = np.concatenate(embeddings)

X_train, X_test, y_train, y_test = train_test_split(
    features,
    df_bert["toxic"],
    test_size=0.25,
    random_state=state,
    stratify=df_bert["toxic"],
)
print(X_train.shape)
print(X_test.shape)
```

#### Логистическая регрессия

```python
param_distributions = {
    "max_iter": [100, 500, 700, 1000],
    "solver": ["lbfgs", "liblinear"],
}

logistic_regression = GridSearchCV(
    estimator=LogisticRegression(random_state=state, class_weight="balanced"),
    param_grid=param_distributions,
    verbose=0,
    n_jobs=-1,
    scoring="f1"
).fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        logistic_regression.best_score_,
        2,
    ),
)
```

#### Случайный лес

```python
param_distributions = {
    "max_depth": [1, 5, 10, 50, 100],
    "n_estimators": [2, 10, 50, 100, 300],
    "min_samples_split": [2, 5, 10],
    "criterion": ["entropy", "gini"],
}


rf = HalvingRandomSearchCV(
    estimator=RandomForestClassifier(random_state=state),
    param_distributions=param_distributions,
    verbose=0,
    n_jobs=-1,
    random_state=state,
    scoring="f1",
    max_resources=1500,
).fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        rf.best_score_,
        2,
    ),
)
```

#### LightGBM

```python
param_distributions = {
    "n_estimators": [10, 50, 70, 100, 150, 200, 500],
    "num_leaves": [2, 5, 10, 30, 50, 100, 150],
    "learning_rate": [0.03, 0.1, 0.2, 0.5, 0.7],
}


LightGBM = HalvingRandomSearchCV(
    estimator=lgbm.LGBMClassifier(random_state=state, force_col_wise=True),
    param_distributions=param_distributions,
    verbose=0,
    n_jobs=-1,
    random_state=state,
    scoring="f1",
    max_resources=2450,
    cv=4,
).fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        LightGBM.best_score_,
        2,
    ),
)
```

#### Проверка на тестовой выборке

Проверим модель с наилучшим значением метрики на тестовой выборке:

```python
print(
    "Метрика f1 на тестовой выборке = ",
    round(f1_score(y_test, rf.best_estimator_.predict(X_test)), 2),
)
```

#### Промежуточный вывод

1. С помощью модели **BERT** провели эмбеддинг признаков
2. Обучили три модели и получили следующие значения метрик *f1*:
    * Логистическая регрессия - 0.85
    * Случайный лес - 0.89
    * LightGBM - 0.83
    
Лучший результат показала модель *Случайный лес*. Далее проверили её на тестовой выборке и получили также хорошее значение метрики ***f1 = 0.89***, что существенно выше целевого значения 0.75.

### TF-IDF

#### Векторизация

Данные уже лемматизированы и очищены от ненужных символов. Произведем векторизацию текста, а перед этим разделим данные на обучающую и тестовую:

```python
df["text"] = df["text"].values.astype("U")
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["toxic"], test_size=0.25, random_state=state
)
print(X_train.shape)
print(X_test.shape)
```

#### Логистическая регрессия

```python
param_distributions = {
    "model__max_iter": [100, 500],
    "model__solver": ["lbfgs", "liblinear"],
}
pipeline = Pipeline(
    [
        # ("vect", CountVectorizer(ngram_range=(1, 4))),
        (
            "tfidf",
            TfidfVectorizer(stop_words=list(set(nltk_stopwords.words("english")))),
        ),
        ("model", LogisticRegression(random_state=state, class_weight="balanced")),
    ]
)
logistic_regression = GridSearchCV(
    pipeline,
    verbose=0,
    n_jobs=-1,
    param_grid=param_distributions,
    scoring="f1",
)
logistic_regression.fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        logistic_regression.best_score_,
        2,
    ),
)
```

#### Случайный лес

```python
param_distributions = {
    "model__criterion": ["entropy", "gini"],
}


pipeline = Pipeline(
    [
        # ("vect", CountVectorizer(ngram_range=(1, 4))),
        (
            "tfidf",
            TfidfVectorizer(stop_words=list(set(nltk_stopwords.words("english")))),
        ),
        ("model", RandomForestClassifier(random_state=state, max_depth=70)),
    ]
)
rf = GridSearchCV(
    pipeline,
    verbose=0,
    n_jobs=-1,
    cv=4,
    param_grid=param_distributions,
    scoring="f1",
)
rf.fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        rf.best_score_,
        2,
    ),
)
```

Очень долго ждать. Решено оставить 70 максимальную глубину, чтобы все таки дождаться хотя бы почти нулевого результата.

#### LightGBM

```python
param_distributions = {
    "model__n_estimators": [100, 200, 500],
    "model__learning_rate": [0.03, 0.1],
}

pipeline = Pipeline(
    [
        # ("vect", CountVectorizer(ngram_range=(1, 4))),
        (
            "tfidf",
            TfidfVectorizer(stop_words=list(set(nltk_stopwords.words("english")))),
        ),
        ("model", lgbm.LGBMClassifier(random_state=state, force_col_wise=True)),
    ]
)
LightGBM = GridSearchCV(
    pipeline,
    verbose=0,
    n_jobs=-1,
    cv=3,
    param_grid=param_distributions,
    scoring="f1",
)
LightGBM.fit(X_train, y_train)

print(
    "Метрика f1 на валидационной выборке = ",
    round(
        LightGBM.best_score_,
        2,
    ),
)
```

#### Проверка на тестовой выборке

```python
print(
    "Метрика f1 на тестовой выборке = ",
    round(f1_score(y_test, LightGBM.best_estimator_.predict(X_test)), 2),
)
```

#### Промежуточный итог

1. С помощью метрики **TF-IDF** векторизовали признаки
2. Обучили три модели и получили следующие значения метрик *f1*:
    * Логистическая регрессия - 0.74
    * Случайный лес - 0.15 (слишком долго обучается)
    * LightGBM - 0.76
    
Лучший результат показала модель *LightGBM*, при этом достигнув целевого значения **0.78** на тестовой выборке.
