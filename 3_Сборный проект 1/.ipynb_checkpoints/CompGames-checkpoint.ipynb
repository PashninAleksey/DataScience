## Анализ продаж компьютерных игр

## Цели проекта

Выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

## Описание данных

Из открытых источников доступны следующие данные о продажах игр: <br>


*Name* — название игры<br>
*Platform* — платформа (например, Xbox или PlayStation)<br>
*Year_of_Release* — год выпуска<br>
*Genre* — жанр игры<br>
*NA_sales* — продажи в Северной Америке (миллионы проданных копий)<br>
*EU_sales* — продажи в Европе (миллионы проданных копий)<br>
*JP_sales* — продажи в Японии (миллионы проданных копий)<br>
*Other_sales* — продажи в других странах (миллионы проданных копий)<br>
*Critic_Score* — оценка критиков (максимум 100)<br>
*User_Score* — оценка пользователей (максимум 10)<br>
*Rating* — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Пропуски и типы данных

Построим столбчатую диаграмму, которая наглядно покажет количество пропусков по столбцам, а затем рассмотрим столбцы с пропусками по отдельности поподробнее:

```python
# пропущенные значения бары
def pass_value_barh(df):
    try:
        plot = (
            (df.isna().sum())
            .to_frame()
            .rename(columns={0: "null_count"})
            .query("null_count > 0")
            .sort_values(by="null_count", ascending=True)
            .plot(kind="barh", figsize=(19, 6), legend=True, fontsize=16)
        )

        plot.set_title(
            "Количество пропусков в столбцах", fontsize=22, color="SteelBlue"
        )
        plot.set_xlabel("Количество пропусков", fontsize=18)
        plot.set_ylabel("Колонки", fontsize=18)

        plt.show()

    except:
        print("пропусков нет или произошла ошибка")
```

#### Столбец 'user_score'

Разберемся со столбцом *user_score*, рассмотрим поподробнее значение *tbd*. Для этого выведем их количество и для примера несколько строк с этим значениеями: 

```python
print(
    "Количество значение " "tbd" ": ",
    df[df["user_score"] == "tbd"]["user_score"].count(),
)
print(
    "Количество пропусков в столбце " "user_score" ": ", df["user_score"].isna().sum()
)
print("20 случайных строк со значением " "tbd" " в столбце " "user_score" ":")
display(df[df["user_score"] == "tbd"].sample(20))
```

При просмотре таблицы со строками со значением "tbd" никакой взаимосвязи не прослеживается. При этом количество строк со значением "tbd" в столбце *user_score* равно **2424**, что весьма существенно, учитывая, что строк у нас всего **16715**, также количество пропусков велико - **6701**. Просто удалить строки с пропусками и значениями "tbd" будет некорректно.

При поиске информации в интернете было найдено,что данная аббревиатура используется при оценке [игр](https://topreytings.ru/tbd-reyting-chto-eto/) и [фильмов](https://www.metacritic.com/faq#item13) и означает, что данный объект еще не оценен. Но так как мы видим, что данная аббревиатура есть и у довольно старых игр, то, видимо, они уже и не будут оценены. Поэтому предлагается поменять значения NaN и tbd на **0**. Но перед этим надо проверить, есть ли вообще оценки **0**, ведь тогда мы испортим статистику данной оценки:

```python
df_backup.loc[df_backup["user_score"] == "0"]
```

Одно значение все таки оказалось равным нулю, это на статистике не отразится. Можем проверить: отзывы у игры про пони, например, на Amazone очень даже приличные, значит здесь какая то ошибка и вместо tbd подкрался 0 в таблицу. Далее меняем значения NaN и tbd на 0

```python
df.loc[df["user_score"] == "tbd", "user_score"] = 0
df["user_score"].fillna(0, inplace=True)
print(
    "Количество значений " "tbd" ": ",
    df[df["user_score"] == "tbd"]["user_score"].count(),
)
print(
    "Количество пропусков в столбце " "user_score" ": ", df["user_score"].isna().sum()
)

# преобразуем тип данных строки 'user_score' в float64
df["user_score"] = pd.to_numeric(df["user_score"])
df.info()
```

## Исследовательский анализ данных

```python
# функция для постройки барплотов
def plt_batplot(
    width=15,
    height=8,
    rotate=30,
    xdata=None,
    ydata=None,
    color="skyblue",
    xlabel="",
    ylabel="",
    title="",
):
    plt.figure(figsize=(width, height))
    plt.xticks(rotation=rotate, horizontalalignment="right")
    ax = sbrn.barplot(x=xdata, y=ydata, color=color)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title);
```

### Сколько игр выпускалось в разные годы

```python
count_games_of_year = df.pivot_table(
    index="year_of_release", values="name", aggfunc="count"
)
count_games_of_year.columns = ["count_games"]
display(count_games_of_year)
```

Довольно большое количество **269** игр после предобработки данных имеют значение года релиза **0**. Посмотрим платформы у которых нулевой год:

Преобладают относительно "древние" приставки, избавимся от нулевых годов:

```python
df = df[df["year_of_release"] != 0]
count_games_of_year = df.pivot_table(
    index="year_of_release", values="name", aggfunc="count"
)
count_games_of_year.columns = ["count_games"]
display(count_games_of_year)
```

Построим столбчатую диаграмму для количества выпущенных игр по годам:

```python
plt_batplot(
    xlabel="Года",
    ylabel="Количество игр",
    title="Количество выпускаемых игр по годам",
    xdata=count_games_of_year.index,
    ydata=count_games_of_year.count_games,
)
```

По графику видим, что количество выпускаемых игр с 1980 по 1990-й год практически не менялось и было очень малым (до 50 штук в год), далее пошел рост быстрыми темпами вплоть до 2008-2009 годов, а затем случился спад практически в два раза. Рост, полагаю, можно объяснить развитием технологий - приставки игровые и компьютеры становились все мощнее и распространённее, спад - сложностью и дороговизной разработки более современных игр.

### Влияние отзывов пользователей и критиков на продажи

В качестве примера выберем платформу X360:

Диаграмма рассеяния для зависимости общей суммы продаж игры от отзывов критиков:

Есть конечно выбросы, но кажется корректным утверждать, что игры которые имели положительные отзывы у критиков, они же и имели успех в продажах.

Диаграмма рассеяния для зависимости общей суммы продаж игры от отзывов пользователей:

```python
# рассмотрим только те случаи у которых есть оценка пользователей
x360[x360["user_score"] > 0].plot(
    kind="scatter", x="user_score", y="sum_sales", ec="blue", figsize=(10, 6)
);
```

В целом схожая ситуация, но больше выбросов, то есть много копий продано, а оценка игры не очень высокая. На то они и пользователи, у каждого свое мнение и зачастую совсем непрофессиональное.

Матрица корреляции для оценок пользователей, оценок критиков и количествa проданных копий игры:

```python
fig, axs = plt.subplots(2, 1, figsize=(15, 10))


ax = sbrn.heatmap(
    x360.loc[(x360["user_score"] > 0), ["user_score", "sum_sales"]]
    .corr()[["sum_sales"]]
    .sort_values(by="sum_sales", ascending=False),
    cmap="BuGn_r",
    annot=True,
    ax=axs[0],
)

ax = sbrn.heatmap(
    x360.loc[(x360["critic_score"] > 0), ["critic_score", "sum_sales"]]
    .corr()[["sum_sales"]]
    .sort_values(by="sum_sales", ascending=False),
    cmap="BuGn_r",
    annot=True,
    ax=axs[1],
);
```

Видим, заметная <b>0,53</b> заметная положительная корреляция между оценкой критиков и количеством проданных копий, и слабая отрицательная корреляции <b>-0,14</b> между оценкой пользователя и количеством проданных копий. Значения бьются с диаграммами рассеивания, которые мы строили до этого.

Чтобы проверить зависимости между оценками критиков и пользователей для других актуальных платформ так же построим матрицы корреляции:

* Влияет ли рейтинг ESRB на продажи в отдельном регионе?

```python
# функция для постройки столбчатых диаграмм
def barplot_show(df_na, df_eu, df_jp, param):
    # столбцчатые диаграммы
    barWidth = 0.3
    r1 = np.arange(len(df_na.index))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    plt.bar(
        r1,
        df_na["na_sales"],
        color="blue",
        width=barWidth,
        edgecolor="black",
        label="Северная Америка",
    )
    plt.bar(
        r2,
        df_eu["eu_sales"],
        color="green",
        width=barWidth,
        edgecolor="black",
        label="Европа",
    )
    plt.bar(
        r3,
        df_jp["jp_sales"],
        color="red",
        width=barWidth,
        edgecolor="black",
        label="Япония",
    )
    plt.xlabel(param, fontweight="bold", fontsize=10)
    plt.ylabel("Продано копий, млн. шт", fontweight="bold", fontsize=10)
    plt.xticks(
        [r + barWidth for r in range(len(df_na.index))],
        df_na.index,
        rotation=45,
        fontsize=14,
    )
    plt.rcParams["figure.figsize"] = [15, 15]
    plt.legend()
    plt.show()
```

```python
# функция для добавления строки "другие"
def other_row(df_in):
    df_tmp = df_in.head().copy()
    df_tmp.loc["others"] = df_in[~df_in.index.isin(df_in.head().index)].sum()
    return df_tmp
```

## Проверка гипотез

Гипотезы следующие:
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки обеих гипотез будем использовать t-тест из модуля SciPy для равнства средних двух независимых генеральных совокупностей.
Определим уровень статистической значимости alpha = 0.05

### Гипотеза 1

Нулевая гипотеза **Н0**  - средние пользовательские рейтинги платформ Xbox One и PC равны;<br>
Альтернативная гипотеза **Н1** - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

```python
# двухсторонняя гипотеза
results = st.ttest_ind(
    actual_data.loc[
        (actual_data["platform"] == "xone") & (actual_data["user_score"] > 0),
        "user_score",
    ],
    actual_data.loc[
        (actual_data["platform"] == "pc") & (actual_data["user_score"] > 0),
        "user_score",
    ],
)

print("p-значение:", results.pvalue)
# уровень статистической значимости
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

При уровне статистической значимости в 0,05 и p-значение=0,1 недостаточно оснований для отклонения 0-ой гипотезы.

### Гипотеза 2

Идем от обратного:<br>
Нулевая гипотеза **Н0**  - Средние пользовательские рейтинги жанров Action и Sports равны;<br>
Альтернативная гипотеза **Н1** - Средние пользовательские рейтинги жанров Action и Sports разные.

```python
# двухсторонняя гипотеза
results = st.ttest_ind(
    actual_data.loc[
        (actual_data["genre"] == "action") & (actual_data["user_score"] > 0),
        "user_score",
    ],
    actual_data.loc[
        (actual_data["genre"] == "sports") & (actual_data["user_score"] > 0),
        "user_score",
    ],
)
print("p-значение:", results.pvalue)
# уровень статистической значимости
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Отклонили нулевую гипотезу. Значит <font color = "blue">остается</font> альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports отличаются между собой.
