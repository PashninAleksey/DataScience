### ceiling_height

```python
#с помощью метода describe() рассмотрим числовое описание данных
data['ceiling_height'].describe()
```

Согласно строительным нормам и правилам (СНиП) высота потолков в жилых помещениях не должна быть меньше 2.5 метров. Но, как видим из описания столбца, минимальная высота аж **1** метр, что очевидно является ошибкой. Проверим есть ли еще в столбце значения меньше 2.5 метров.

```python
#срез данных, в которых высота потолка ниже допустимых 2 метров
print(data.query('ceiling_height<2'))
data.query('ceiling_height<2')['ceiling_height'].hist();
```

Таких случаев у нас 3 штуки. Другие данные в этих строках впорядке. Полагаю, случилась какая-то опечатка или иная ошибка. Можно задать вместо этих аномально маленьких значений значение None и в дальнейшем заполнить их медианными значениями высот потолков соответствующих классу данного жилья (*эконом, премиум, бизнес, элитный*).

```python
data.loc[data['ceiling_height']<2,'ceiling_height'] = None
#при проверке выдает Empty DataFrame, значит значений меньше 2 метров больше нет
#print(data.query('ceiling_height<2'))
```

Из числового описания всей таблицы известно, что есть аномальная высота потолка **100** метров. Необходимо проверить, есть ли еще нереалистично высокие значения и исправить их или избавиться при невозможности исправить.

```python
#срез данных, в которых высота потолка выше 5 метров
print(data.query('ceiling_height>5'))
data.query('ceiling_height>5')['ceiling_height'].hist();
```

Видно, что имеется большое количество значений от **25** метров и выше, можно предположить, что при внесении этих данных произошла ошибка, а именно, не поставили точку разделитель дробной части, то есть эти значения надо поделить на **10**.

```python
#если значение в столбце выше или равно 25, то делим на 10
def set_new_height(x,threshold):
    if x>=threshold:
        x=x/10
    return x
data['ceiling_height'] = data['ceiling_height'].apply(lambda x: set_new_height(x,25))
print(data.query('ceiling_height>5'))
print("Количество квартир с высотой потолка выше 5 метров =", len(data.query('ceiling_height>5')))
```

Осталось 17 квартир с аномально высокими потолками.<br>
Так же, как и для аномально низких потолков зададим этим данным значение None, и в дальнейшем заполним медианными значениями высот соответствующего класса жилья.


```python
data.loc[data['ceiling_height']>5,'ceiling_height'] = None
#при проверке выдает Empty DataFrame, значит значений больше 5 метров нет
#print(data.query('ceiling_height>5'))
```

Необходимо классифицировать недвижимость по стоимости и заменить пропуски в *сeiling_height* медианными значениями высот потолка в соответствующем классе.<br>
В статьях про рынок недвижимости часто приводят цифры, говорящие о том, что цена жилья *бизнес*-класса выше *эконом*-класса в 1.5-2 раза, *премиум* - дороже *эконома* в 2.5-3 раза, *элитное*, соответственно, еще дороже.

```python
data['last_price'].describe()
```

Можно предположить, что большинство недвижимости это *эконом* жилье. То есть за среднюю цену *эконом* жилья можно взять медианное значение **4.65** млн. рублей. Значит можно определить примерную классификацию следующим образом:

|    Класс  | Цена, млн     | 
|:-----     |:-------- 
|Эконом     |   до 9  
|Бизнес     |   от 9 до 15 
|Премиум    |   от 15 до 20
|Элитное    |   от 20

Добавим колонку классификации к датафрейму: 

```python
def realty_type(row):
    price = row['last_price']
    classication = 'economy'
    if price>=9000000 and price<15000000:
        classication = 'business'
    elif price>=15000000 and price<20000000:
        classication = 'premium'
    elif price>20000000:
        classication = 'elite'
    return classication
data['realty_type'] = data.apply(realty_type, axis=1)
data.sample(5)
```

Колонка добавлена. Теперь заполним пропуски для высот потолков.

```python
# посмотрим на медиану в каждой категории
data.groupby(by='realty_type')['ceiling_height'].median()
```

Данные соответствуют действительности.

```python
# заполняем пропуски медианным значением
# количество пропусков до
print(data['ceiling_height'].isnull().sum())
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('realty_type')['ceiling_height']
        .transform('median'))
# проверяем количество пропусков после заполнения 
print(data['ceiling_height'].isnull().sum())
```

Колонка с высотой потолка полностью обработана от пропусков и аномальных значений.

Теперь, если вывести на экран наименования населенных пунктов, то можно заметить неявные дубликаты. 

```python
# количество уникальных наименований
print(data['locality_name'].value_counts().count())

data['locality_name'].value_counts().sample(50)
```

Количество уникальных наименований на данный момент **364**.<br>
Причинами этих дубликатов являются: 
1. Буквы "е" и "ё" в слове "поселок" и посёлок".
2. Дополнительные слова "коттеджный", "городского типа", "городской"...
3. Наличие слова поселок или его отсутствие (в случае с Мурино).

Предлагаю добавить дополнительный столбец, в котором будут только имена собственные без типа населенного пункта.

```python
def naming(row):
    name = row['locality_name']
    index=0
    #найдем индекс большой буквы, начиная с которой мы возьмем топоним
    for i in str(name):
        if i.isupper():
            break
        else:
            index+=1
    return name[index:]
    
    
data['proper_name'] = data.apply(naming, axis=1)
print(data['proper_name'].value_counts())
```

Количество уникальных названий оказалось не **364**, а **305**. Хороший результат!

### Изучение и описание параметров в объявлениях

Построим для следующих параметров коробчатые диаграммы, на их основе удалим выбросы, затем построим гистограмму и посмотрим числовые характеристики для числовых столбцов<br>
  - общая площадь;
  - жилая площадь;
  - площадь кухни;
  - цена объекта;
  - количество комнат;
  - высота потолков;
  - этаж квартиры;
  - тип этажа квартиры («первый», «последний», «другой»);
  - общее количество этажей в доме;
  - расстояние до центра города в метрах;
  - расстояние до ближайшего аэропорта;
  - расстояние до ближайшего парка;
  - день и месяц публикации объявления

```python
#функция для постойки boxplot
def for_box(col):
    print('Коробчатая диаграмма для столбца ' + col)
    sbrn.boxplot(x=col, data = data)
    plt.show()
    print()
    
#функция для постойки histogram
def for_hist(col,b=10):
    print('Гистограмма для столбца ',col)
    data.hist(column=col,bins = b);
    plt.show()
    print()
    
#функция для отображения числовых характеристик
def for_describe(col):
    print('Числовое описание столбца ' + col + ":")
    print(data[col].describe())
    print()
    print('Количество пропусков в столбце ' + col +":", data[col].isnull().sum())
    print()
    
```

### Какие факторы больше всего влияют на общую (полную) стоимость объекта

Построим диаграммы рассеивания и найдем значение корреляции между стоимостью жилья и следующими параметрами:
1. жилая площадь;
2. площадь кухни;
3. количество комнат;
4. тип этажа, на котором расположена квартира (первый, последний, другой);
5. дата размещения (день недели, месяц, год).

```python
#напишем функцию для вывода диаграммы рассеивания и значения корреляции для площади жилья, кухни и количества комнат
def diag_and_corr(col):
    data.plot(x='last_price', y=col, kind='scatter', grid = True) 
    plt.title('Зависимость стоимости жилья от ' + col, fontsize= 12)
    plt.show()
    print("Корреляция между стоимостью и " + col, data['last_price']. corr(data[col]))
```

```python
#формула для постройки графиков зависимости стоимости от времени публикации
def plotting_last_price(col):
    df = data.pivot_table(
    index=col, 
    values='last_price', 
    aggfunc='mean')
    df['last_price'] /= 10**6
    display(df)
    ax = df.plot(xlabel='Год', y='last_price', kind = 'line', style='o-', grid = True)
    ax.set_ylabel('last_price, 10**6 руб.')
    
```

```python
plotting_last_price('month_exposition')
plt.title('Зависимость цены от месяца размещения объявления', fontsize= 12);
```

**От недели:**

```python
plotting_last_price('weekday_exposition')
plt.title('Зависимость цены от дня недели размещения объявления', fontsize= 12);
```
