# Прогноз температуры стали

## Описание задачи

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Необходимо построить модель, которая предскажет температуру стали.

## Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

## Описание данных

Данные состоят из файлов, полученных из разных источников:
* **data_arc_new.csv** — данные об электродах;
* **data_bulk_new.csv** — данные о подаче сыпучих материалов (объём);
* **data_bulk_time_new.csv** — данные о подаче сыпучих материалов (время);
* **data_gas_new.csv** — данные о продувке сплава газом;
* **data_temp_new.csv** — результаты измерения температуры;
* **data_wire_new.csv** — данные о проволочных материалах (объём);
* **data_wire_time_new.csv** — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. <br>
В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

Целевым признаком является последняя измеренная температура.<br>
Объектом моделирования является ковш (строки с одинаковым значением ***key***)<br>
Целевая метрика: ***MAE***<=6.8

## Предобработка данных

### Агрегирование данных

Агрегируем данные в таблицах по партиям. 

Таблицы данных о подаче сыпучих материалов (время) и данных о проволочных материалах (время) нас не интересуют.

#### Данные об электродах

 В новой таблице будут номер партии, время нагрева и вместо двух мощностей будет только полная мощность, которая высчитывается по следующей формуле: <br>
$S$=$\sqrt{P^2+Q^2}$        
где: 
- S - полная мощность,
- Q - реактивная мощность,
- P - активная мощность

Сперва добавим колонку с полной мощностью:

```python
data_arc["Полная мощность"] = (
    data_arc["Реактивная мощность"] ** 2 + data_arc["Активная мощность"] ** 2
) ** 0.5
data_arc.head()
```

Теперь добавим столбец с временем нагрева (для универсальности в секундах):

```python
data_arc["Длительность нагрева"] = (
    (data_arc["Конец нагрева дугой"] - data_arc["Начало нагрева дугой"])
    .dt.total_seconds()
    .astype(int)
)
data_arc.head(10)
```

Агрегируем необходимые данные:

```python
df_arc = data_arc[["key", "Длительность нагрева", "Полная мощность"]].pivot_table(
    index=["key"], aggfunc="sum"
)
df_arc.head()
```

#### Данные о подаче сыпучих материалов (объём)

Вспомним как выглядела таблица:

```python
data_bulk
```

Агрегируем на всякий случай данные по *key*:

```python
df_bulk = data_bulk.pivot_table(index=["key"], aggfunc="sum")
df_bulk
```

Видимо, можно было обойтись без суммирования, так как все значения *key* оказались уникальные.

#### Данные о проволочных материалах (объём)

Снова оказалось, что в таблице все значения *key* были уникальными.

## Обучение моделей

Обучим 4 регрессионные модели:
1. ***LinearRegression***
2. ***RandomForestRegressor***
3. ***LGBMRegressor***
4. ***CatBoostRegressor***

Для начала создадим кросс-валидатор для наших тренировочных данных и метрику оценки:

```python
kfold = KFold(n_splits=5)
scoring = "neg_mean_absolute_error"
print("done...")
```

### Линейная регрессия

С помощью кросс-валидации определим лучшее значение метрики MAE для линейной регрессии:

```python
print(
    "MAE = ",
    abs(cross_val_score(
        LinearRegression(), X_train, y_train, cv=kfold, scoring=scoring
    )).mean()
)
```

Получили результат на тренировочной выборке **MAE=7.61**

### Случайный лес

```python
param_grid = {
    "max_depth": [1, 5, 10, 50, 100],
    "n_estimators": [2, 10, 50, 100],
    "min_samples_split": [2, 5, 10],
}

rf = GridSearchCV(
    estimator=RandomForestRegressor(random_state=RANDOM_STATE),
    param_grid=param_grid,
    verbose=0,
    n_jobs=-1,
    cv=kfold,
    scoring=scoring,
).fit(X_train, y_train)

print("MAE = ", abs(rf.best_score_))
```

Получили результат **MAE=7.14**

### LightGBM

```python
param_grid = {
    "n_estimators": [2, 5, 10, 70, 100, 200, 300],
    "num_leaves": [2, 5, 10, 30, 50, 150],
    "learning_rate": [0.1, 0.2, 0.5, 0.7],
}

LightGBM = GridSearchCV(
    estimator=lgbm.LGBMRegressor(random_state=RANDOM_STATE, force_row_wise=True),
    param_grid=param_grid,
    verbose=0,
    n_jobs=-1,
    cv=kfold,
    scoring=scoring,
).fit(X_train, y_train)

print("MAE = ", abs(LightGBM.best_score_))
```

Получили результат **MAE=7.11**

### CatBoost

```python
param_grid = {
    "learning_rate": [0.03, 0.7, 0.9],
    "iterations": [100, 200, 300, 500],
}

CatBoost = GridSearchCV(
    estimator=CatBoostRegressor(
        loss_function="MAE",
    ),
    param_grid=param_grid,
    verbose=0,
    n_jobs=-1,
    cv=kfold,
    scoring=scoring,
).fit(X_train, y_train)

print("MAE = ", abs(CatBoost.best_score_))
```

Получили результат **MAE=6.98**

Подобранные гиперпараметры лучшей модели:

```python
CatBoost.best_estimator_.get_params()
```

### Проверка на тестовой выборке

Наилучшее значение метрики на обучающей выборке показала модель **CatBoost** (*MAE*=6.98)

Проверим работу данной модели на тестовой выборке:

```python
print("MAE = ", mean_absolute_error(y_test, CatBoost.best_estimator_.predict(X_test)))
```

Получили значение метрики **MAE** на тестовой выборке у модели CatBoost равное **6.53**

После того, как мы выбрали модель, обеспечивающую лучшее значение метрики, проведем анализ важности факторов:

```python
feature_importance = CatBoost.best_estimator_.feature_importances_
sorted_idx = np.argsort(feature_importance)
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align="center")
plt.yticks(
    range(len(sorted_idx)),
    np.array(df.drop(["Температура"], axis=1).columns)[sorted_idx],
)
plt.title("Feature Importance");
```

Получили, что самый весомый вклад в прогнозирование конечной температуры вносит признак "Длительность нагрева", что звучит очень логично.
